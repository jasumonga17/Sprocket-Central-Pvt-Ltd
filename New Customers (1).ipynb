```python
# Importing  Necessary Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("/Users/jasdeepmonga/Library/Containers/com.microsoft.Excel/Data/Downloads/NewCustomers.csv",header=1)
data.head()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
#Dropping Unecessary Columns
data.drop(['Unnamed: 16','Unnamed: 17','Unnamed: 18','Unnamed: 19','Unnamed: 20'],axis=1,inplace=True)
```

```python
# Job Title
data['job_title']=data['job_title'].fillna("Other")
```

```python
#Job Industry
data['job_industry_category']=data['job_industry_category'].fillna("Other")
```

```python
#Gender-Undisclosed category
data['gender']=data['gender'].fillna("U")
```

```python
# LastName
data['last_name']=data['last_name'].fillna("None")
```

```python
data.head()
```

```python
data.info()
```

```python
# Changing datatype of DOB
data['DOB']=pd.to_datetime(data['DOB'])
```

```python
# Calculating 'Age' of the Customer based on DOB

from datetime import date,datetime

now=datetime.now()

data['Age']=now.year-data['DOB'].dt.year
```

```python
# Reading the Dataset
data.head()
```

```python
# State Purchases
m_state=data.groupby(['state'])['past_3_years_bike_related_purchases'].sum().reset_index().sort_values(by='past_3_years_bike_related_purchases',ascending=False)
m_state.head()
```

```python
# Job Title
j_title=data.groupby(['job_title'])['past_3_years_bike_related_purchases'].sum().reset_index().sort_values(by='past_3_years_bike_related_purchases',ascending=False)
j_title.head()
```

```python
# Job Industry Category
j_category=data.groupby(['job_industry_category'])['past_3_years_bike_related_purchases'].sum().reset_index().sort_values(by='past_3_years_bike_related_purchases',ascending=False)
j_category.head()
```

```python
# Wealth Segment
w_segment=data.groupby(['wealth_segment'])['past_3_years_bike_related_purchases'].sum().reset_index().sort_values(by='past_3_years_bike_related_purchases',ascending=False)
w_segment.head()
```

```python
#gender
gender=data.groupby(['gender'])['past_3_years_bike_related_purchases'].sum().reset_index().sort_values(by='past_3_years_bike_related_purchases',ascending=False)
gender.head()
```

```python
# Age
age=data.groupby(['Age'])['past_3_years_bike_related_purchases'].sum().reset_index().sort_values(by='past_3_years_bike_related_purchases',ascending=False)
age.head()
```

```python
# Owns a Car
car=data.groupby(['owns_car'])['past_3_years_bike_related_purchases'].sum().reset_index().sort_values(by='past_3_years_bike_related_purchases',ascending=False)
car.head()
```

```python
plt.figure(figsize=(5,7))
sns.barplot(x='owns_car',y='past_3_years_bike_related_purchases',data=car)
plt.show()
```
